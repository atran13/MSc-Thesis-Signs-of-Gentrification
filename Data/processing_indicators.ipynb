```python
import pandas as pd
```

```python
indicators = pd.read_csv('indicators_2020.csv', sep=';')
indicators
```

```python
indicators.columns = ['neighborhood','type of region','code','population (count)',
                      'age 15-25 (count)','age 25-45 (count)','age 45-65 (count)','age >65 (count)',
                      'Western migrants (count)','non-Western migrants (count)','property value (x1000 euro)',
                      'education low (count)','education secondary (count)','education high (count)',
                      'income per inhabitant (x1000 euro)','40% with lowest income (%)','20% with highest income (%)']
indicators
```

```python
indicators['age 15-25 (%)'] = round(100 * indicators['age 15-25 (count)'] / indicators['population (count)'], 2)
indicators['age 25-45 (%)'] = round(100 * indicators['age 25-45 (count)'] / indicators['population (count)'], 2)
indicators['age 45-65 (%)'] = round(100 * indicators['age 45-65 (count)'] / indicators['population (count)'], 2)
indicators['age >65 (%)'] = round(100 * indicators['age >65 (count)'] / indicators['population (count)'], 2)
```

```python
indicators['migrants (%)'] = round(100 * (indicators['Western migrants (count)'] + indicators['non-Western migrants (count)']) / indicators['population (count)'], 2)
indicators['Western migrants (%)'] = round(100 * indicators['Western migrants (count)'] / indicators['population (count)'], 2)
indicators['non-Western migrants (%)'] = round(100 * indicators['non-Western migrants (count)'] / indicators['population (count)'], 2)
```

```python
indicators['education low (count)'] = indicators['education low (count)'].str.extract(pat='(\d+)', expand=False).astype('Int64')
indicators['education secondary (count)'] = indicators['education secondary (count)'].str.extract(pat='(\d+)', expand=False).astype('Int64')
indicators['education high (count)'] = indicators['education high (count)'].str.extract(pat='(\d+)', expand=False).astype('Int64')

indicators['education low (%)'] = round(100 * indicators['education low (count)'] / indicators['population (count)'], 2)
indicators['education secondary (%)'] = round(100 * indicators['education secondary (count)'] / indicators['population (count)'], 2)
indicators['education high (%)'] = round(100 * indicators['education high (count)'] / indicators['population (count)'], 2)
```

```python
mod_indicators_2020 = indicators[['neighborhood', 'type of region', 'code', 'population (count)',
                                  'age 15-25 (%)', 'age 25-45 (%)', 'age 45-65 (%)', 'age >65 (%)',
                                  'migrants (%)', 'Western migrants (%)', 'non-Western migrants (%)',
                                  'education low (%)', 'education secondary (%)', 'education high (%)',
                                  'property value (x1000 euro)', 'income per inhabitant (x1000 euro)',
                                  '40% with lowest income (%)', '20% with highest income (%)']]
mod_indicators_2020
```

```python
mod_indicators_2020['property value (x1000 euro)'] = mod_indicators_2020['property value (x1000 euro)'].str.extract(pat='([0-9]+[,./]*[0-9]*)', expand=False).astype('float')
mod_indicators_2020['income per inhabitant (x1000 euro)'] = mod_indicators_2020['income per inhabitant (x1000 euro)'].str.extract(pat='([0-9]+[,./]*[0-9]*)', expand=False).astype('float')
mod_indicators_2020['40% with lowest income (%)'] = mod_indicators_2020['40% with lowest income (%)'].str.extract(pat='([0-9]+[,./]*[0-9]*)', expand=False).astype('float')
mod_indicators_2020['20% with highest income (%)'] = mod_indicators_2020['20% with highest income (%)'].str.extract(pat='([0-9]+[,./]*[0-9]*)', expand=False).astype('float')
```
